```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
mta_stations = pd.read_csv('data/MTA_Stations.csv')
mta_stations.head()
```

```python
mta_stations = mta_stations[['Complex ID','Stop Name','Station ID','Borough']]
mta_stations
```

```python
mta_rides = pd.read_csv('data/MTA_Passengers.csv')
mta_rides.head()
```

```python
mta_rides = mta_rides[['Day of Week','Hour of Day','Estimated Average Ridership','Origin Station Complex ID','Destination Station Complex ID']]
mta_rides
```

```python
G = nx.MultiDiGraph
```

```python
G = nx.from_pandas_edgelist(mta_rides, source='Origin Station Complex ID', target='Destination Station Complex ID', 
                     edge_attr= True,
                     create_using=G, edge_key=None)
```

```python
for _, row in mta_stations.iterrows():
    complex_id = row['Complex ID']
    attributes = row.to_dict()
    G.add_node(complex_id, **attributes)
```

### Prompt

Once you’re able to represent this data, as part of graph-based representation, you are
expected to develop queries on the graph-based representation which should return the
following:
1) What are the top 5 origin sub-way stations from where most riders took subway ride:
    1. a. Across each borough (Manhattan, Brooklyn, Queens, Bronx, State Island)
    2. b. On Monday, Tuesday, Wednesday (combined)
    3. c. On Saturday and Sunday (combined)
    4. d. Between 1am-5am across all days and boroughs
    5. e. Between 6am-9am across all days and boroughs
2) What are the top 5 destination sub-way stations from where most riders took subway ride:
    1. a. Across each borough (Manhattan, Brooklyn, Queens, Bronx, State Island)
    2. b. On Thursday and Friday (combined)
    3. c. On Saturday only
    4. d. Between 12am-5am across all days and boroughs
    5. e. Between 6pm-9pm across all days and boroughs
3) What are the top 10 congested source-destination sub-way stations pair on:
    1. Monday between 1pm-2pm?
    2. On Queens borough, on Fridays between 6pm-9pm?
    3. On Brooklyn borough, Ridership between 1am-5am?
    4. Source is Brooklyn, Destination is Manhattan, Monday-Thursday 6am-7am?
    5. Source is Bronx, Destination is Manhattan, Monday-Thursday 6am-7am?
    6. Source is Staten Island, Destination is Manhattan, Monday-Thursday 6am-7am?

### 1a:

```python
avg_ridership_per_station = {}

for node in G.nodes:
    edges = G.edges(node, data=True)
    
    riderships = [data.get('Estimated Average Ridership', 0) for _, _, data in edges]
    if riderships:
        avg_ridership_per_station[node] = sum(riderships)
    else:
        avg_ridership_per_station[node] = 0
        
stations_by_borough = {}

for node, data in G.nodes(data=True):
    borough = data.get('Borough')
    if borough not in stations_by_borough:
        stations_by_borough[borough] = []
    stations_by_borough[borough].append((node, avg_ridership_per_station[node]))
    
top_stations_per_borough = {}

for borough, stations in stations_by_borough.items():
    top_stations = sorted(stations, key=lambda x: x[1], reverse=True)[:5]
    top_stations_per_borough[borough] = top_stations
    
for borough, stations in top_stations_per_borough.items():
    print('Top 5 stations in '+ borough + ':')
    for station, ridership in stations:
        print("Station ID: " + str(station))
    print()
```

### 1b:

```python
days = {'Monday', 'Tuesday', 'Wednesday'}

total_ridership_per_station = {}

for node in G.nodes:
    edges = G.edges(node, data=True)
    riderships = [
        data.get('Estimated Average Ridership', 0) for _, _, data in edges if data.get('Day of Week') in days]
    total_ridership_per_station[node] = sum(riderships)
top_5_stations = sorted(total_ridership_per_station.items(), key=lambda x: x[1],reverse=True)[:5]
for station, ridership in top_5_stations:
    print('Station ID: ' + str(station))
```

### 2a:

```python
ridership_per_station = {}

for _, destination, data in G.edges(data=True):
    ridership = data.get('Estimated Average Ridership', 0)
    
    if destination in ridership_per_station:
        ridership_per_station[destination] += ridership
    else:
        ridership_per_station[destination] = ridership
        
stations_by_borough = {}

for node, data in G.nodes(data=True):
    borough = data.get('Borough')
    if borough not in stations_by_borough:
        stations_by_borough[borough] = []
    ridership = ridership_per_station.get(node, 0)
    stations_by_borough[borough].append((node, ridership))
    
top_stations_per_borough = {}

for borough, stations in stations_by_borough.items():
    top_stations = sorted(stations, key=lambda x: x[1], reverse=True)[:5]
    top_stations_per_borough[borough] = top_stations
    
for borough, stations in top_stations_per_borough.items():
    print('Top 5 stations in '+ borough + ':')
    for station, ridership in stations:
        print("Station ID: " + str(station))
    print()
```

### 3a:

```python
congested_pairs_monday_1pm_2pm = []

for origin, destination, data in G.edges(data=True):
    ridership = data.get('Estimated Average Ridership', 0)
    day_of_week = data.get('Day of Week')
    hour = data.get('Hour of Day')

    if day_of_week == 'Monday' and hour in [13,14]:
        congested_pairs_monday_1pm_2pm.append(((origin, destination), ridership))

top_10_monday_1pm_2pm = sorted(congested_pairs_monday_1pm_2pm, key=lambda x: x[1],  reverse=True)[:10]
for (origin, destination), ridership in top_10_monday_1pm_2pm:
    print('Origin Station ID: ' + str(origin) + ', Destination: ' + str(destination))
```

### 3b:

```python
congested_pairs_fri_queens_6_9 = defaultdict(int)

for origin, destination, data in G.edges(data=True):
    ridership = data.get('Estimated Average Ridership', 0)
    day_of_week = data.get('Day of Week')
    hour = data.get('Hour of Day')

    if day_of_week == 'Friday' and hour in [18,19,20,21]:
        origin_borough = G.nodes[origin].get('Borough')
        destination_borough = G.nodes[destination].get('Borough')

        if origin_borough == 'Q' and destination_borough == 'Q':
            congested_pairs_fri_queens_6_9[(origin, destination)] += ridership

        
top_10_congested_pairs_fri_queens_6_9 = sorted(congested_pairs_fri_queens_6_9.items(), key=lambda x: x[1],  reverse=True)[:10]
for (origin, destination), ridership in top_10_congested_pairs_fri_queens_6_9:
    print('Origin Station ID: ' + str(origin) + ', Destination: ' + str(destination))
```
